# Welche ted Talks generieren die meisten Kommentare?

```python
ted["comments_per_view"] = ted["comments"] / ted["views"]
```

```python
ted.sort_values("comments_per_view", ascending = False).head(3)
```

```python
ted["views_per_comment"] = ted["views"] / ted["comments"]
```

```python
ted.sort_values("views_per_comment", ascending = True).head(3)
```

# Verteilungen der Anzahl Kommentare Visualisieren

```python
ted[["comments"]].plot(kind = "hist", bins = 20)
plt.show()
```

```python
ted[ted["comments"] < 1500][["comments"]].plot(kind = "hist", bins = 30)
plt.show()
```
